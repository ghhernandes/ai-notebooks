```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn import linear_model
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
dataset = '/Users/hernandes/datasets/heart_failure_clinical_records_dataset.csv'
df = pd.read_csv(dataset)
```

```python
df.head()
```

## Death vs Age

```python
fig, ax = plt.subplots(1, 2, figsize=(13, 7))
sns.boxplot(x='DEATH_EVENT', y='age', data=df, ax=ax[0])
sns.stripplot(x='DEATH_EVENT', y='age', data=df, ax=ax[1])
```

```python
sns.catplot(x='DEATH_EVENT', y='age', data=df, kind='violin')
```
